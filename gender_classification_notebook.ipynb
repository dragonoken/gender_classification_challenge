Let's try training machine learning algorithms on this dataset

```python
# Split the dataset
from sklearn.model_selection import train_test_split
X = gender_df.drop('Gender', axis=1)
y = gender_df['Gender']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

```python
# Import machine learning algorithms
from sklearn.linear_model import LogisticRegression

from sklearn.svm import SVC
from sklearn.grid_search import GridSearchCV

from sklearn.linear_model import Perceptron
```

```python
# Create classifiers
logreg = LogisticRegression()
gridSVC = GridSearchCV(SVC(), param_grid={'C':[1, 10, 100, 1000], 'gamma':[1, .1, .01, .001]})
perceptron = Perceptron()
```

```python
# Fit them to the training dataset
logreg.fit(X_train, y_train)
gridSVC.fit(X_train, y_train)
perceptron.fit(X_train, y_train)
```

```python
# Test them
pred_logreg = logreg.predict(X_test)
pred_grid = gridSVC.predict(X_test)
pred_perceptron = perceptron.predict(X_test)

from sklearn.metrics import confusion_matrix, classification_report

print("Logistic Regression")
print(confusion_matrix(y_test, pred_logreg))
print(classification_report(y_test, pred_logreg))
print()

print("Support Vector Machine")
print(confusion_matrix(y_test, pred_grid))
print(classification_report(y_test, pred_grid))
print()

print("Perceptron")
print(confusion_matrix(y_test, pred_perceptron))
print(classification_report(y_test, pred_perceptron))
print()
```
